###Handling Null Values in dataset

```python
df.isnull().sum()
```

```python
# Check for missing values
missing_values = df.isna()
print(missing_values)
```

```python
# Replace null values with either mean, median or mode
df['culmen_length_mm'].fillna(df['culmen_length_mm'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['culmen_depth_mm'].fillna(df['culmen_depth_mm'].mean(),inplace=True)
df['flipper_length_mm'].fillna(df['flipper_length_mm'].mean(),inplace=True)
df['body_mass_g'].fillna(df['body_mass_g'].mean(),inplace=True)
```

###Here target is Species of penguin

The dependent variable is 'species' and the independent variables are 'island','culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g','sex'

```python
df.head()
```

Checking the correlation of independent variables with the target

First we have to encode the categorical columns to obtain the accurate correlation result.

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.species=le.fit_transform(df.species)
```

```python
df.island=le.fit_transform(df.island)
```

```python
df.sex=le.fit_transform(df.sex)
```

```python
df.head()
```

Now checking correlation of independent variables with the target (species)

```python
df.corr().species.sort_values(ascending=False)
```

<h1> So we have first distinguished between the Dependent and Independent Variables which are as follows:

1. 'species' of the penguin to be predicted.
2. all rest of the categories are the predictors.

<h1> We have also done the encoding of the categorical columns, including 'sex','species' and 'island'.</h1>
