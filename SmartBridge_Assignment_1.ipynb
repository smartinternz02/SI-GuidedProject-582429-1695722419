Now, we will make a dictionary of some random data of Laptops, in which we will insert 5 observations with its 5 features.


```python
d={"name":["ASUS Vivobook Pro 15","HP Laptop 15","Dell G15-5520","Samsung Galaxy Book2(NP750)","Lenovo Ideapad Slim 3"],"ram":[16,8,8,4,np.nan],"rom":[256,512,np.nan,1024,256],"screen_size":[14.0,15.6,13.8,np.nan,15.6],"price":[65990,49980,57490,61000,45500]}
d
```

Above shown is our dictionary with randome values, along with some 'null' values.

Now, next step is to convert this dictionary into a DataFrame model with the help of Pandas library of Python.

```python
df=pd.DataFrame(d)
df
```

NOTE:

If we want to insert random floating point values using numPy into our Pandas DataFrame, then we can do it using the following code:



```python
num_rows = 4
num_columns = 4
df2 = pd.DataFrame(index=range(num_rows), columns=range(num_columns))

# Generate random values using NumPy and assign them to the DataFrame
for i in range(num_rows):
    for j in range(num_columns):
        df2.at[i, j] = np.random.random()

df2
```
